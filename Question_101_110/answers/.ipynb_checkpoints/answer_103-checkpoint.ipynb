## 打开、显示、保存图像

基于python脚本语言开发的数字图片处理包，比如PIL,Pillow, opencv, scikit-image等。PIL和Pillow只提供最基础的数字图像处理，功能有限。其中Pillow是由PIL发展而来的（即Pillow是由PIL fork而来，因此还是要从PIL中进行import）；opencv实际上是一个c++库，只是提供了python接口，更新速度非常慢。scikit-image是基于scipy的一款图像处理包，它将图片作为numpy数组进行处理，正好与matlab一样。这里选择Skimage模块进行数字图像处理。

### 1 使用opencv-python

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
file_dir = r"../../Tutorial/images/imori.jpg"
```

```python
# 读取, cv2的读取和存储格式是BGR
img = cv2.imread(file_dir) 
print(img.shape)
print(type(img))
# 显示
plt.imshow(img)
plt.axis('off')
plt.show()
# 保存
cv2.imwrite('./imori.jpg', img)
```

### 2 使用PIL

```python
from PIL import Image
img = Image.open(file_dir)
print(img.size) # 可以用size获取长宽
print(img.mode)
print(type(img))  # 得到的是PIL对象格式
plt.imshow(img)
# 关闭轴信息
plt.axis("off")
plt.show()

# 图片保存,这行代码不仅能保存图片，还可以转换格式
img.save('./imori.jpg')
```

### 3 使用scikit-image

```python
from skimage import io
#skimage读取图片, 读取和存储格式是RGB
img=io.imread(file_dir, as_gray=False)
print(img.shape)
print(type(img))
# 显示
plt.imshow(img)
plt.axis('off')
plt.show()
# 图片保存
io.imsave('./imori.jpg', img)
```
